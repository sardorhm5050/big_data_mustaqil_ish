## **6-qadam. 2024 yil energiya taqsimoti grafigi**
Energiya manbalarining ulushini pasta diagramma va bar grafik orqali vizuallashtiramiz

```python
print("\n=== 2024 YIL ENERGIYA TAQSIMOTI GRAFIGI ===")

# Asosiy energiya manbalari uchun ma'lumotlar
main_sources = df_energy_2024.iloc[1:11]  # Asosiy manbalarni olamiz

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# 1. Pasta diagramma
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F', '#BB8FCE', '#85C1E9']
wedges, texts, autotexts = ax1.pie(main_sources['Ulush'],
                                  labels=main_sources['Energiya_Manbai'],
                                  autopct='%1.1f%%',
                                  colors=colors,
                                  startangle=90)

ax1.set_title('2024 Yil Global Energiya Taqsimoti', fontsize=16, fontweight='bold', pad=20)

# 2. Bar grafik
bars = ax2.barh(main_sources['Energiya_Manbai'], main_sources['Ishlab_Chiqarish_TWh'],
               color=colors, edgecolor='black', linewidth=0.5)

ax2.set_title('Energiya Ishlab Chiqarish Hajmi (TWh)', fontsize=16, fontweight='bold')
ax2.set_xlabel('Terawatt-soat (TWh)', fontsize=12)
ax2.grid(axis='x', alpha=0.3)

# Bar grafikdagi qiymatlarni yozish
for bar in bars:
    width = bar.get_width()
    ax2.text(width + 1000, bar.get_y() + bar.get_height()/2,
             f'{width:,.0f}', ha='left', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

## **7-qadam. Tarixiy o'zgarish grafigi**
1965 va 2024 yillardagi energiya ishlab chiqarish hajmlarini solishtiruvchi bar grafik chizamiz

```python
print("\n=== ENERGIYA MANBALARINING TARIXIY O'ZGARISHI GRAFIGI ===")

plt.figure(figsize=(14, 8))

# Bar grafik
x = np.arange(len(df_historical))
width = 0.35

bars1 = plt.bar(x - width/2, df_historical['1965_TWh'], width,
               label='1965 yil', color='lightblue', edgecolor='grey')
bars2 = plt.bar(x + width/2, df_historical['2024_TWh'], width,
               label='2024 yil', color='lightcoral', edgecolor='grey')

plt.xlabel('Energiya Manbalari', fontsize=12, fontweight='bold')
plt.ylabel('Ishlab Chiqarish (TWh)', fontsize=12, fontweight='bold')
plt.title('Energiya Manbalarining 1965-2024 Yillardagi O\'zgarishi',
          fontsize=16, fontweight='bold', pad=20)
plt.xticks(x, df_historical['Energiya_Manbai'], rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Qiymatlarni yozish
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 500,
                f'{height:,.0f}', ha='center', va='bottom', fontsize=8,
                fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
print("\n=== QAYTA TIKLANUVCHI ENERGIYA O'SISHI GRAFIGI ===")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 1. O'sish foizlari
growth_rates = [6940, 171600, 928, 328, 48, 219]
sources = df_renewable['Energiya_Manbai']

bars = ax1.bar(sources, growth_rates, color=['#FF9999', '#66B2FF', '#99FF99', '#FFB366', '#FF99FF', '#FFFF99'])
ax1.set_title('Qayta Tiklanuvchi Energiya O\'sishi (2000-2024)', fontsize=14, fontweight='bold')
ax1.set_ylabel('O\'sish Foizi (%)', fontsize=12)
ax1.tick_params(axis='x', rotation=45)
ax1.set_yscale('log')  # Logarifmik shkala

# Qiymatlarni yozish
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height * 1.1,
            f'{height:,.0f}%', ha='center', va='bottom', fontsize=9,
            fontweight='bold', rotation=45)

# 2. Hajmlar taqqoslash
years = ['2000 yil', '2024 yil']
x_pos = np.arange(len(sources))
width = 0.35

bars1 = ax2.bar(x_pos - width/2, df_renewable['2000_TWh'], width,
               label='2000 yil', color='lightgreen')
bars2 = ax2.bar(x_pos + width/2, df_renewable['2024_TWh'], width,
               label='2024 yil', color='darkgreen')

ax2.set_title('Energiya Ishlab Chiqarish Hajmi (TWh)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Terawatt-soat (TWh)', fontsize=12)
ax2.set_xticks(x_pos)
ax2.set_xticklabels(sources, rotation=45, ha='right')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
